```python
import pandas as pd

# 1. 전체 출력 옵션 설정 (요약 내용이 길어도 줄바꿈 없이 모두 출력)
pd.set_option('display.max_colwidth', None)  # 열 너비 무제한
pd.set_option('display.max_rows', 1000)      # 필요한 만큼 행 보여주기
pd.set_option('display.max_columns', None)   # 모든 컬럼 표시

# 2. CSV 불러오기 (파일 경로 수정)
file_path = "school_info_refined_with_summary_20250512_212705.csv"  # ← 파일명에 맞게 수정하세요
df = pd.read_csv(file_path)

# 3. summary 컬럼 선택
summary_columns = [col for col in df.columns if col.startswith("summary_")]

# 4. '대학명(영문)' + summary 컬럼만 추출
filtered_df = df[['대학명(영문)'] + summary_columns]

# 5. summary 컬럼 중 하나라도 비어있지 않은 행만 필터링
filtered_df = filtered_df[filtered_df[summary_columns].notnull().any(axis=1)]

# 6. 결과 출력
filtered_df.reset_index(drop=True, inplace=True)
filtered_df
```
