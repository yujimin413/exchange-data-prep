```python
import pandas as pd

# 원본 및 변경 파일 경로
origin_path = 'reports_origin.csv'
changed_path = 'reports_with_only_changed_20250513_230904.csv'

# 데이터 불러오기
df_origin = pd.read_csv(origin_path, index_col=0)
df_changed = pd.read_csv(changed_path, index_col=0)

# 원본 복사본 만들어 두기 (덮어쓰기 전 상태 저장)
df_before_update = df_origin.copy()

# 덮어쓰기 수행
df_origin.update(df_changed)

# 변경된 행만 필터링 (행 단위 비교)
mask_changed_rows = (df_origin != df_before_update).any(axis=1)
df_modified_rows = df_origin[mask_changed_rows]

# 덮어씌워진 행만 확인
print(df_modified_rows)

# 필요한 경우 파일로 저장
df_modified_rows.to_csv('modified_rows_only.csv', encoding='utf-8-sig')
```

```python
import pandas as pd

# 파일 경로
origin_path = 'reports_origin.csv'
changed_path = 'reports_with_only_changed_20250513_230904.csv'

# 데이터 불러오기
df_origin = pd.read_csv(origin_path, index_col=0)
df_changed = pd.read_csv(changed_path, index_col=0)

# 덮어쓰기 전 복사본 보존
df_before_update = df_origin.copy()

# 덮어쓰기 수행
df_origin.update(df_changed)

# 변경 내역 확인
changed_details = []

for idx in df_changed.index:
    if idx in df_before_update.index:
        for col in df_changed.columns:
            before = df_before_update.at[idx, col]
            after = df_origin.at[idx, col]
            if before != after:
                changed_details.append({
                    'index': idx,
                    'column': col,
                    'before': before,
                    'after': after
                })

# DataFrame으로 보기 좋게 출력
df_changes = pd.DataFrame(changed_details)

# 출력 또는 저장
print(df_changes)
df_changes.to_csv('changed_columns_only.csv', index=False, encoding='utf-8-sig')
```
