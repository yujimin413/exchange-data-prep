```python
import pandas as pd

# íŒŒì¼ ê²½ë¡œ
school_summary_path = 'school_info_refined_with_summary_20250514_000416.csv'
school_merged_path = 'reports_merged.csv'
school_cleaned_path = 'reports_cleaned.csv'

# íŒŒì¼ ë¡œë”©
df_summary = pd.read_csv(school_summary_path)
df_merged = pd.read_csv(school_merged_path)
df_cleaned = pd.read_csv(school_cleaned_path)

# ëŒ€í•™ëª…(ì˜ë¬¸) ì»¬ëŸ¼ ì •ì œ
def get_clean_name_set(df):
    return set(df['ëŒ€í•™ëª…(ì˜ë¬¸)'].dropna().astype(str).str.strip().unique())

names_summary = get_clean_name_set(df_summary)
names_merged = get_clean_name_set(df_merged)
names_cleaned = get_clean_name_set(df_cleaned)

# ê° ë¹„êµ ê²°ê³¼
only_in_summary_vs_merged = names_summary - names_merged
only_in_summary_vs_cleaned = names_summary - names_cleaned

# ë¹„êµ ê²°ê³¼ ì¶œë ¥
print("âœ… summaryì—ë§Œ ìˆê³  mergedì—ëŠ” ì—†ëŠ” ëŒ€í•™ëª… ìˆ˜:", len(only_in_summary_vs_merged))
print("âœ… summaryì—ë§Œ ìˆê³  cleanedì—ëŠ” ì—†ëŠ” ëŒ€í•™ëª… ìˆ˜:", len(only_in_summary_vs_cleaned))

# ë‘ ì§‘í•©ì´ ë™ì¼í•œì§€ í™•ì¸
if only_in_summary_vs_merged == only_in_summary_vs_cleaned:
    print("\nğŸ¯ ë‘ ì°¨ì§‘í•©ì€ ë™ì¼í•©ë‹ˆë‹¤.")
else:
    print("\nâš ï¸ ë‘ ì°¨ì§‘í•©ì€ ë‹¤ë¦…ë‹ˆë‹¤.")

    diff_1 = sorted(only_in_summary_vs_merged - only_in_summary_vs_cleaned)
    diff_2 = sorted(only_in_summary_vs_cleaned - only_in_summary_vs_merged)

    if diff_1:
        print("\nğŸ“Œ mergedì—ëŠ” í¬í•¨ë˜ì§€ë§Œ cleanedì—ëŠ” ì—†ëŠ” ì°¨ì´:")
        print(diff_1)

    if diff_2:
        print("\nğŸ“Œ cleanedì—ëŠ” í¬í•¨ë˜ì§€ë§Œ mergedì—ëŠ” ì—†ëŠ” ì°¨ì´:")
        print(diff_2)
```
